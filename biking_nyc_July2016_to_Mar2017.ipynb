### The column names in October 2017 change. To be able to concatenate the dataframes together I need to change the data frame column names to match the new October 2017 columns.

```python
# The columns in OCT 2016 change from lowercase to upper case with spaces
# I need to rename the column headings in the old dataframe (concat_02) to the new format.

concat_02 = concat_02.rename(columns={'tripduration':'Trip Duration', 'startime':'Start Time', \
                         'stoptime':'Stop Time', 'start station id':'Start Station ID', \
                         'start station name':'Start Station Name', \
                         'start station latitude':'Start Station Latitude', \
                         'start station longitude':'Start Station Longitude', \
                         'end station id':'End Station ID', \
                         'end station name':'End Station Name', \
                         'end station latitude':'End Station Latitude', \
                         'end station longitude':'End Station Longitude', \
                         'bikeid':'Bike ID', \
                         'usertype':'User Type', \
                         'birth year':'Birth Year', \
                         'gender':'Gender'})
# check to see if columns are retitled
concat_02.head()
```
