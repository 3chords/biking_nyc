```python
# use concat to append the rows of the dfs
concat_30 = pd.concat([df_2019_05, df_2019_06], sort=True)
# reset the index of the new DF
concat_30.reset_index()
concat_30.head()
```

```python
# read files into dataframe
df_2019_07 = pd.read_csv(bike_2019_07)
df_2019_07.head()
```

```python
# get # of rows
df_1907_rows = df_2019_07.count()
print(df_1907_rows)
```

```python
# use concat to append the files
concat_31 = pd.concat([concat_30, df_2019_07], sort=True)
# reset the index of the new DF
concat_31.reset_index()
concat_31.head()
```

```python
# get # of rows for concatenate
concat_31_rows = concat_31.count()
print(concat_31_rows)
```

```python
# Rename the columns to be consistent with the expect
# I need to rename the column headings in the old dataframe to the new format.

concat_31 = concat_31.rename(columns={'tripduration':'Trip Duration', 'startime':'Start Time', \
                         'stoptime':'Stop Time', 'start station id':'Start Station ID', \
                         'start station name':'Start Station Name', \
                         'start station latitude':'Start Station Latitude', \
                         'start station longitude':'Start Station Longitude', \
                         'end station id':'End Station ID', \
                         'end station name':'End Station Name', \
                         'end station latitude':'End Station Latitude', \
                         'end station longitude':'End Station Longitude', \
                         'bikeid':'Bike ID', \
                         'usertype':'User Type', \
                         'birth year':'Birth Year', \
                         'gender':'Gender'})
# check to see if columns are retitled
concat_31.head()
```
